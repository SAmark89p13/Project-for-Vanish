# Importar las librerias y la tabla que hicimos antes del anterior entregable 

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
#Así guarde la tabla en la anterior entrega para iniciar este proceso
df=pd.read_csv("ventas_filtradas.csv")
```

# Eliminar columnas que tienen todos los valores nulos

```python
df_cleaned = df.dropna(axis=1, how="all")
df_cleaned.info()
```

# Histograma de ventas totales

```python
plt.figure(figsize=(14, 8))
sns.boxplot(data=df_cleaned, x="REGION", y="TOTAL_VALUE_SALES", palette="pastel")
plt.title("Boxplot de Valor Total de Ventas por Marca")
plt.xlabel("REGION")
plt.ylabel("Total Value Sales")
plt.xticks(rotation=90)
plt.savefig("1Boxplot-VALOR-VENTAS-POR-MARCA.png", dpi=300, bbox_inches="tight")
plt.show()
```

# Analizar la tendencia de ventas en el tiempo:

        

## Tendencia de ventas por región

```python
# Agrupamos los datos para fechas y ventas para tener un dataframe específico
df_region= df_cleaned.groupby(["YEAR_WEEK","REGION"]).agg({
    "TOTAL_UNIT_SALES": "sum",
    "TOTAL_VALUE_SALES": "sum"
}).reset_index()
# Ordenamos por fechas YEAR WEEK
df_region = df_region.sort_values("YEAR_WEEK")
df_region.head()
```

```python
#Definimos el tamaño de figura y creamos nuestro gráfico lineplot
plt.figure(figsize=(16,10))
sns.lineplot(
    data=df_region,
    x='YEAR_WEEK',
    y='TOTAL_UNIT_SALES',
    hue= 'REGION',
    marker='o'
)
# Agregamos titulos y etiquetas
plt.title("Tendencia de Unidades Vendidas a lo largo del tiempo por Región")
plt.xlabel("Año-Semana")
plt.ylabel("Total Unit Sales")
plt.xticks(rotation=90)
plt.savefig("3Tendencia de Unidades Vendidas a lo largo del tiempo por Región.png", dpi=300, bbox_inches="tight")
plt.show()
```

## En la gráfica de arriba Y en la de abajo podemos observar como el áerea con más ventas es auto scannning México y después es el área 6. La que menos vende es el área 4

```python
sns. heatmap(pd.crosstab(df[ "CATEGORY_cat" ] , df [ "REGION"], values = df [ "TOTAL_UNIT_SALES" ], aggfunc = "mean"), annot = True,cmap="coolwarm")
plt.savefig("4Mapa de calor más ventas por región.png", dpi=300, bbox_inches="tight")
```

## Tendencia de ventas por marca

```python
# Agrupamos los datos para fechas y ventas para tener un dataframe específico
df_marca= df_cleaned.groupby(["YEAR_WEEK","BRAND"]).agg({
    "TOTAL_UNIT_SALES": "sum",
    "TOTAL_VALUE_SALES": "sum"
}).reset_index()
# Ordenamos por fechas YEAR WEEK
df_marca = df_marca.sort_values("YEAR_WEEK")
df_marca.head()
```

```python
#Definimos el tamaño de figura y creamos nuestro gráfico lineplot

plt.figure(figsize=(16,10))
sns.lineplot(
    data=df_marca,
    x='YEAR_WEEK',
    y='TOTAL_VALUE_SALES',
    hue='BRAND',
    marker='o'
)
#Obtener la semana en la que se alcazaron valor de ventas totales mayor o igual a 100k
meta_week = df_marca.loc[df_marca["TOTAL_VALUE_SALES"] >= 100000, "YEAR_WEEK"].iloc[0]

#LÍNEA DE META PARA 100K 
plt.axvline(x=meta_week, color='red', linestyle='--', linewidth=2, label=f'Semana Meta: {meta_week}')

# Agregamos titulos y etiquetas
plt.title("Tendencia de Unidades Vendidas a lo largo del tiempo por Marca")
plt.xlabel("Año-Semana")
plt.ylabel("Total Value Sales")
plt.xticks(rotation=90)
plt.tight_layout()
plt.legend(loc="upper left", bbox_to_anchor=(2, 2))
plt.savefig("5Tendencia de Unidades Vendidas a lo largo del tiempo por Marca.png", dpi=300, bbox_inches="tight")

plt.show()
```

## En la gráfica de arriba podemos observar como la marca con más ventas de unidades es Cloralex y HARPIC. La que menos vende es Cuellos, puntos y machitas (incluso podría ser que no hay datos o hay datos faltantes ya que no se ni una subida)
## Además podemos observar como la semna 22 del 2023 fue la mejor semana de ventas

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crear el scatter plot
ax = sns.scatterplot(
    data=df_cleaned,
    x="TOTAL_UNIT_SALES",
    y="TOTAL_VALUE_SALES",
    hue="REGION",
    palette="coolwarm"
)
# Crear el regretion plot, línea de regresión
sns.regplot(
    data=df_cleaned,
    x="TOTAL_UNIT_SALES",
    y="TOTAL_VALUE_SALES",
    scatter=False,    # Evita que se vuelvan a dibujar los puntos
    ax=ax,            # Dibujar la línea en el mismo eje
    color="red",    # Color de la línea de regresión
    line_kws={"linewidth": 2}  # Ancho de la línea
)

# Mover la leyenda fuera del gráfico
ax.legend(loc="upper left", bbox_to_anchor=(1, 1)) 

# Personalizar el gráfico
plt.title("Relación entre Unidades vendidas y Valor Total de Ventas por Región")
plt.xlabel("Total Unit Sales")
plt.ylabel("Total Value Sales")
plt.savefig("6Relación entre Unidades vendidas y Valor Total de Ventas por Región.png", dpi=300, bbox_inches="tight")
plt.show()
```

# En el gráfico de arriba observamos como hay una correlación entre más unidades vendidas más valor de ventas totales existe.
