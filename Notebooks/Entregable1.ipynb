```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Cargar los archivos 
hoja_no1 = pd.read_excel("calendardim.xlsx", sheet_name=None)
df_hoja_no2 = pd.read_csv("categorydim.csv")
hoja_no3= pd.read_excel("DIM_PRODUCT (1).xlsx", sheet_name=None)
hoja_no4 = pd.read_excel("DIM_SEGMENT (1).xlsx", sheet_name=None)
df_hoja_no5 = pd.read_csv("FACT_SALES (1).csv")
#Inspeccionar los archivos que son un diccionario
df_hoja_no1 = hoja_no1["Sheet1"]
df_hoja_no3=  hoja_no3["Sheet1"]
df_hoja_no4=  hoja_no4["SEGMENT"]
##Tomar información de cada uno 
df_hoja_no1.info()
df_hoja_no2.info()
df_hoja_no3.info()
df_hoja_no4.info()
df_hoja_no5.info()
```

# Borrar los duplicados 

```python
df_calendar = df_hoja_no1.drop_duplicates(keep=False) 
df_category = df_hoja_no2.drop_duplicates(keep=False) 
df_product = df_hoja_no3.drop_duplicates(keep=False) 
df_segment= df_hoja_no4.drop_duplicates(keep=False) 
df_sales = df_hoja_no5.drop_duplicates(keep=False) 
```

# Limpiar los datas frames

```python
df_calendar.columns=df_calendar.columns.str.strip().str.upper()
df_calendar['DATE'] = pd.to_datetime(df_calendar['DATE'], dayfirst=True, errors='coerce')
df_calendar.columns = df_calendar.columns.str.strip().str.upper()
df_sales
```

```python
# Normalizar nombres de columnas
df_category.columns = df_category.columns.str.strip().str.upper()
df_category["CATEGORY"] = df_category["CATEGORY"].astype(str).str.strip().str.upper()

df_category.info()
```

```python
df_segment.columns = df_segment.columns.str.strip().str.upper()
#Probar limpiar con un bucle 
# Para todas las columnas de interés, se limpia el texto
for col in ["SEGMENT", "CATEGORY", "ATTR1", "ATTR2", "ATTR3", "FORMAT"]:
    if col in df_segment.columns:
        df_segment[col] = df_segment[col].astype(str).str.strip().str.upper()

df_segment.info()
```

```python
for col in ["PRODUCT_ID", "CATEGORY", "SEGMENT"]:
    if col in df_sales.columns and df_sales[col].dtype == object:
        df_sales[col] = df_sales[col].str.strip().str.upper()
df_sales.info()
```

# Empezar a unir los data flrames 

```python
 # Crear una columna compuesta en cada DataFrame
df_calendar['YEAR_WEEK'] = df_calendar['WEEK'].astype(str) + '-' + df_calendar['WEEK_NUMBER'].astype(str)
df_sales['YEAR_WEEK'] = df_sales['WEEK'].astype(str)
# Realizar la unión basada en la clave compuesta
df_sales_calendar = pd.merge(df_sales, df_calendar, on='YEAR_WEEK', how='left')
df_product
```

```python
df_product.rename(columns={"ITEM": "ITEM_CODE"}, inplace=True)
df_sales_product_calendar = pd.merge(df_sales_calendar, df_product, on='ITEM', how='left')
df_product
```
